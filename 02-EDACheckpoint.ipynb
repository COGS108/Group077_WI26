# Research Question

We will test whether **estimated caffeine consumption per capita (by country)** is associated with **country-level health and social outcomes**, after accounting for major confounders.

Primary question: Across countries, is higher estimated caffeine intake associated with differences in mental health and mortality indicators (e.g., depression burden and all-cause mortality)?

Secondary questions:
- Do associations differ depending on the dominant caffeine source (coffee vs tea vs carbonated/energy drinks)?
- Do associations remain after controlling for factors like GDP per capita, smoking prevalence, alcohol use, and other relevant confounders?

This is an **inference** project using regression-based modeling and sensitivity analyses to test robustness to uncertainty in the caffeine estimates.


## Background and Prior Work

Caffeine is one of the most widely consumed psychoactive substances worldwide, typically consumed through coffee, tea, and other caffeinated beverages. Because caffeine affects sleep, mood, and cardiovascular physiology, population-level differences in caffeine intake could plausibly relate to mental health and mortality patterns.

At the individual level, observational research has reported associations between caffeine intake and depression-related outcomes. For example, analyses using NHANES data examined caffeine consumption and depression measures, suggesting a relationship may exist, though causality is not established and confounding remains a major concern.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Similarly, large observational analyses have examined caffeine intake and all-cause or cause-specific mortality, often identifying non-linear associations and emphasizing careful interpretation and adjustment for confounders.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

However, most prior work is within-country and individual-level, while our project is between-country and ecological. This makes the question different: we are not claiming caffeine causes outcomes in individuals, but asking whether country-level patterns co-vary and whether those patterns persist after controlling for structural factors (wealth, health system capacity, smoking, alcohol use, etc.).

To estimate country-level caffeine intake, we will start from publicly available per-capita consumption estimates for coffee and tea by country for recent years.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) We will then convert these amounts to approximate caffeine intake using caffeine-content references (acknowledging large variation across preparation methods and products).<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5)

In addition, beverage-market composition data (percent of caffeinated beverage volume sales by category) can help characterize where caffeine is coming from in different countries (coffee vs tea vs carbonates vs energy drinks), which we can use for subgroup and interaction analyses.<a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6)

# Hypothesis


We hypothesize that countries with higher estimated caffeine consumption per capita will show measurable differences in mental health and mortality indicators, though the direction may differ by outcome. Based on prior observational findings, we predict a weak-to-moderate association between higher caffeine intake and lower depression burden and/or lower all-cause mortality, but we expect substantial confounding by economic development and lifestyle factors.<a name="cite_ref-1b"></a>[<sup>1</sup>](#cite_note-1)

We also hypothesize that the dominant caffeine source will matter: countries where caffeine comes mainly from energy drinks or carbonated beverages may show different patterns than countries where caffeine primarily comes from coffee or tea.<a name="cite_ref-6b"></a>[<sup>6</sup>](#cite_note-6)


## Data

### Data overview

Our project uses seven country-level datasets to estimate caffeine exposure and compare it with health and social outcomes. The final dataset combines a **country-year panel** for outcomes and confounders with a **country-level beverage dataset** for coffee and tea consumption.

The two beverage datasets are recent cross-sectional measures. We use them to build an approximate measure of caffeine intake per capita. The health and confounder datasets are country-year panels, which allow us to examine patterns across multiple years.

#### Dataset 1: Coffee consumption per capita
This dataset comes from World Population Review and reports coffee consumption by country in **kilograms per capita per year** for **2023**. After extracting the table from the downloaded HTML file, we keep the country name and coffee consumption columns, convert the consumption values to numeric form, remove missing values, remove duplicate country rows, and sort the table from highest to lowest consumption.

- **Source:** World Population Review, Coffee Consumption by Country  
- **Unit:** kg per capita per year  
- **Year used:** 2023  
- **Variables used:** `country`, `coffee_kg_per_capita`  
- **Purpose in the project:** Forms one part of our estimated caffeine exposure measure  

#### Dataset 2: Tea consumption per capita
This dataset also comes from World Population Review and reports tea consumption by country in **kilograms per capita per year** for **2022**. We clean it in the same way as the coffee dataset so the two can be merged by country.

- **Source:** World Population Review, Tea Consumption by Country  
- **Unit:** kg per capita per year  
- **Year used:** 2022  
- **Variables used:** `country`, `tea_kg_per_capita`  
- **Purpose in the project:** Forms the second part of our estimated caffeine exposure measure  

#### Dataset 3: Depression prevalence
This dataset is a country-year table loaded from our manually downloaded IHME / GBD export. It provides the depression measure used in our analysis.

- **Source:** IHME / GBD export  
- **Unit:** country-level depression measure (stored as `depression_pct`)  
- **Variables used:** `country`, `year`, `depression_pct`  
- **Purpose in the project:** Main mental health outcome  

#### Dataset 4: All-cause mortality
This dataset is another country-year table loaded from our IHME / GBD export. It provides the mortality outcome used in our analysis.

- **Source:** IHME / GBD export  
- **Unit:** rate per 100,000 people  
- **Variables used:** `country`, `year`, `mortality_per_100k`  
- **Purpose in the project:** Main mortality outcome  

#### Dataset 5: GDP per capita
This dataset comes from Our World in Data and is used as a major confounder because national wealth is likely related to both beverage consumption patterns and health outcomes.

- **Source:** Our World in Data / World Bank  
- **Unit:** PPP-adjusted GDP per capita  
- **Variables used:** `country`, `iso3`, `year`, `gdp_per_capita`  
- **Purpose in the project:** Economic confounder  

#### Dataset 6: Smoking prevalence
This dataset comes from Our World in Data and gives the share of adults who smoke.

- **Source:** Our World in Data / WHO  
- **Unit:** percent of adults  
- **Variables used:** `country`, `iso3`, `year`, `smoking_pct`  
- **Purpose in the project:** Lifestyle confounder  

#### Dataset 7: Alcohol consumption
This dataset comes from Our World in Data and reports alcohol consumption per adult.

- **Source:** Our World in Data / WHO  
- **Unit:** liters of pure alcohol per adult per year  
- **Variables used:** `country`, `iso3`, `year`, `alcohol_liters`  
- **Purpose in the project:** Lifestyle confounder  

#### Wrangling and merge process
We first clean the coffee and tea datasets separately, standardize their column names, and save them as processed CSV files. We then load the health and confounder datasets, standardize their key columns, and merge them into a country-year panel using country and year. After that, we merge the beverage data onto the panel by country.

Using the cleaned coffee and tea data, we create two derived variables:
- `est_caffeine_mg_day`, our estimated daily caffeine intake per capita
- `dominant_source`, which labels each country as coffee-dominant or tea-dominant

We also save two final outputs for later analysis:
- `merged_panel.csv`, which contains the full country-year panel
- `merged_cross_section.csv`, which contains the most recent cross-section used in several EDA plots

#### Limitations
This is an ecological dataset built from country-level aggregates, so it cannot be used to make claims about individual people. In addition, our caffeine estimate is approximate because caffeine content varies by preparation method and product. A second limitation is that the coffee and tea inputs are recent cross-sectional measurements, while the health and confounder data span multiple years. This means our estimated caffeine exposure is treated as relatively stable over time, which is a simplifying assumption rather than a direct year-by-year measure.

### Dataset 1 — Coffee Consumption per Capita (2023)

This dataset reports **country-level coffee consumption rates** measured in **kg per capita per year** for 2023. We scrape it from World Population Review, standardize column names, convert to numeric, flag outliers via IQR, and save the cleaned version.

**Key concerns:** Ecological aggregate data; definitions may vary across sources; does not directly measure caffeine mg.

```python
from pathlib import Path
import re
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display

import subprocess, importlib
for pkg in ['lxml', 'html5lib']:
    try:
        importlib.import_module(pkg)
    except ImportError:
        subprocess.check_call(['pip', 'install', '-q', pkg])

RAW       = Path('data/00-raw')
INTERIM   = Path('data/01-interim')
PROCESSED = Path('data/02-processed')
FIG       = Path('figures')
for p in [RAW, INTERIM, PROCESSED, FIG]:
    p.mkdir(parents=True, exist_ok=True)

coffee_html = RAW / 'wpr_coffee.html'
try:
    all_tables = pd.read_html(coffee_html, flavor='lxml')
except ImportError:
    all_tables = pd.read_html(coffee_html, flavor='html.parser')

coffee_raw = next(
    (t for t in all_tables
     if any('kg' in str(c).lower() and 'capita' in str(c).lower() for c in t.columns)),
    all_tables[0]
)

coffee_raw.columns = [
    re.sub(r'\W+', '_', c.strip()).strip('_').lower() for c in coffee_raw.columns
]

cname = next((c for c in coffee_raw.columns if 'country' in c), coffee_raw.columns[0])
crate = next((c for c in coffee_raw.columns if 'kg' in c and 'capita' in c), coffee_raw.columns[1])

coffee_df = (
    coffee_raw[[cname, crate]]
    .rename(columns={cname: 'country', crate: 'coffee_kg_per_capita'})
    .assign(coffee_kg_per_capita=lambda d: pd.to_numeric(d['coffee_kg_per_capita'], errors='coerce'))
    .dropna()
    .drop_duplicates(subset='country')
    .sort_values('coffee_kg_per_capita', ascending=False)
    .reset_index(drop=True)
)

print(f"Coffee: {coffee_df.shape[0]} countries  |  "
      f"min={coffee_df['coffee_kg_per_capita'].min():.2f}  "
      f"median={coffee_df['coffee_kg_per_capita'].median():.2f}  "
      f"max={coffee_df['coffee_kg_per_capita'].max():.2f} kg/cap")

coffee_df.to_csv(PROCESSED / 'coffee_processed.csv', index=False)
display(coffee_df.head(10))
```

### Dataset 2 — Tea Consumption per Capita (2022)

Same approach as coffee — scraped from World Population Review. Reports **kg per capita per year** for 2022.

**Key concerns:** "Tea" may include culturally distinct beverages (e.g., mate); ecological measurement; no direct caffeine measurement.

```python
tea_html = RAW / 'wpr_tea.html'
try:
    all_tables_tea = pd.read_html(tea_html, flavor='lxml')
except ImportError:
    all_tables_tea = pd.read_html(tea_html, flavor='html.parser')

tea_raw = next(
    (t for t in all_tables_tea
     if any('kg' in str(c).lower() and 'capita' in str(c).lower() for c in t.columns)),
    all_tables_tea[0]
)

tea_raw.columns = [
    re.sub(r'\W+', '_', c.strip()).strip('_').lower() for c in tea_raw.columns
]

tname = next((c for c in tea_raw.columns if 'country' in c), tea_raw.columns[0])
trate = next((c for c in tea_raw.columns if 'kg' in c and 'capita' in c), tea_raw.columns[1])

tea_df = (
    tea_raw[[tname, trate]]
    .rename(columns={tname: 'country', trate: 'tea_kg_per_capita'})
    .assign(tea_kg_per_capita=lambda d: pd.to_numeric(d['tea_kg_per_capita'], errors='coerce'))
    .dropna()
    .drop_duplicates(subset='country')
    .sort_values('tea_kg_per_capita', ascending=False)
    .reset_index(drop=True)
)

print(f"Tea: {tea_df.shape[0]} countries  |  "
      f"min={tea_df['tea_kg_per_capita'].min():.2f}  "
      f"median={tea_df['tea_kg_per_capita'].median():.2f}  "
      f"max={tea_df['tea_kg_per_capita'].max():.2f} kg/cap")

tea_df.to_csv(PROCESSED / 'tea_processed.csv', index=False)
display(tea_df.head(10))
```

### Datasets 3–7 — Our World In Data (OWID) Health Outcomes and Confounders (Country-Year Panels)

We download CSV exports from Our World in Data (OWID) for:
- **Depression prevalence** (age-standardized, IHME GBD)
- **All-cause mortality rate** (age-standardized, IHME GBD)
- **GDP per capita** (PPP, World Bank)
- **Smoking prevalence** (% adults, WHO)
- **Alcohol consumption** (liters per adult, WHO)

All OWID CSVs follow the same format: `Entity, Code, Year, <metric>`. We write a reusable loader to standardize them.

```python

def load_owid_csv(filepath, value_keyword, rename_to):
    """Load an OWID-format CSV (Entity, Code, Year, <value>)."""
    df = pd.read_csv(filepath)
    val_cols = [c for c in df.columns if value_keyword.lower() in c.lower()]
    if not val_cols:
        standard = {'Entity', 'Code', 'Year', 'Day'}
        non_standard = [c for c in df.columns if c not in standard]
        val_col = non_standard[-1] if non_standard else df.columns[-1]
        print(f"    Note: '{value_keyword}' not found. Using '{val_col}'")
    else:
        val_col = val_cols[0]
    
    col_map = {val_col: rename_to}
    if 'Entity' in df.columns: col_map['Entity'] = 'country'
    if 'Code' in df.columns: col_map['Code'] = 'iso3'
    if 'Year' in df.columns: col_map['Year'] = 'year'
    df = df.rename(columns=col_map)
    
    keep = [c for c in ['country', 'iso3', 'year', rename_to] if c in df.columns]
    df = df[keep].copy()
    df[rename_to] = pd.to_numeric(df[rename_to], errors='coerce')
    if 'iso3' in df.columns:
        df = df.dropna(subset=['iso3'])
        df = df[df['iso3'].str.len() == 3]
    df = df.dropna(subset=[rename_to])
    return df


def load_ihme_csv(filepath, rename_to):
    """Load an IHME GHDx-format CSV (location_name, year, val, etc.)."""
    df = pd.read_csv(filepath)
    print(f"    IHME format. Shape: {df.shape}")
    
    col_map = {}
    if 'location_name' in df.columns: col_map['location_name'] = 'country'
    if 'val' in df.columns: col_map['val'] = rename_to
    df = df.rename(columns=col_map)
    
    keep = [c for c in ['country', 'year', rename_to] if c in df.columns]
    df = df[keep].copy()
    df[rename_to] = pd.to_numeric(df[rename_to], errors='coerce')
    df = df.dropna(subset=['country', rename_to])
    
    # Remove regional aggregates
    region_keywords = ['Global', 'World', 'income', 'region', 'SDI',
                       'North Africa', 'Sub-Saharan', 'Latin America',
                       'Southeast Asia', 'Central Europe', 'High-income',
                       'Low-income', 'Middle-income', 'Oceania',
                       'East Asia', 'South Asia', 'Western Europe',
                       'Eastern Europe', 'Caribbean', 'Andean',
                       'Central Asia', 'Tropical', 'Southern',
                       'Australasia', 'super-region']
    mask = ~df['country'].str.contains('|'.join(region_keywords), case=False, na=False)
    df = df[mask]
    return df


def detect_and_load(filepath, value_keyword, rename_to):
    """Auto-detect OWID vs IHME format and load accordingly."""
    peek = pd.read_csv(filepath, nrows=2)
    cols = list(peek.columns)
    
    if 'Entity' in cols and 'Code' in cols:
        print(f"    Format: OWID")
        return load_owid_csv(filepath, value_keyword, rename_to)
    elif 'location_name' in cols and 'val' in cols:
        print(f"    Format: IHME GHDx")
        return load_ihme_csv(filepath, rename_to)
    else:
        print(f"    Format: unknown. Columns: {cols[:8]}")
        return load_owid_csv(filepath, value_keyword, rename_to)


owid_sources = {
    'depression': {'file': 'owid_depression.csv', 'keyword': 'depressive', 'rename': 'depression_pct'},
    'mortality':  {'file': 'owid_mortality.csv',   'keyword': 'death',      'rename': 'mortality_per_100k'},
    'gdp':        {'file': 'owid_gdp.csv',         'keyword': 'GDP per',    'rename': 'gdp_per_capita'},
    'smoking':    {'file': 'owid_smoking.csv',      'keyword': 'smoke',      'rename': 'smoking_pct'},
    'alcohol':    {'file': 'owid_alcohol.csv',      'keyword': 'Alcohol',    'rename': 'alcohol_liters'},
}

owid_dfs = {}
for key, info in owid_sources.items():
    fpath = RAW / info['file']
    if not fpath.exists():
        print(f"{key}: not found")
        continue
    
    with open(fpath, 'r', encoding='utf-8', errors='ignore') as f:
        first_line = f.readline().strip()
    if first_line.startswith('<!') or '<html' in first_line.lower():
        print(f"{key}: file is HTML, not CSV — re-download")
        continue
    
    try:
        owid_dfs[key] = detect_and_load(fpath, info['keyword'], info['rename'])
        tmp = owid_dfs[key]
        print(f"{key}: {tmp.shape[0]} rows, {tmp['country'].nunique()} countries, "
              f"years {tmp['year'].min():.0f}–{tmp['year'].max():.0f}")
    except Exception as e:
        print(f"{key}: failed — {e}")

print(f"\nLoaded {len(owid_dfs)}/{len(owid_sources)} OWID datasets")
if len(owid_dfs) < len(owid_sources):
    missing = set(owid_sources.keys()) - set(owid_dfs.keys())
    print(f"Missing: {missing}")
```

```python
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

panel = None
for key in ['depression', 'mortality', 'gdp', 'smoking', 'alcohol']:
    if key not in owid_dfs:
        continue
    df_ow = owid_dfs[key].copy()
    if panel is None:
        panel = df_ow
    else:
        panel = panel.merge(
            df_ow.drop(columns='iso3', errors='ignore'),
            on=['country', 'year'], how='outer')

if panel is not None and 'iso3' in panel.columns:
    panel = panel[panel['iso3'].notna() & (panel['iso3'].str.len() == 3)]

print(f"Panel (outcomes+confounders): {panel.shape[0]} rows, "
      f"{panel['country'].nunique()} countries, "
      f"years {panel['year'].min():.0f}–{panel['year'].max():.0f}")

PROCESSED = Path('data/02-processed')

try:
    beverage = coffee_df[['country', 'coffee_kg_per_capita']].merge(
        tea_df[['country', 'tea_kg_per_capita']], on='country', how='inner')
    print("Beverage data loaded from memory (coffee_df, tea_df)")
except NameError:
    coffee_df = pd.read_csv(PROCESSED / 'coffee_processed.csv')
    tea_df = pd.read_csv(PROCESSED / 'tea_processed.csv')
    beverage = coffee_df[['country', 'coffee_kg_per_capita']].merge(
        tea_df[['country', 'tea_kg_per_capita']], on='country', how='inner')
    print("Beverage data loaded from CSV files")

COFFEE_FACTOR = 32.9  # 1 kg/yr ≈ 2.74 g/day × 12 mg/g
TEA_FACTOR    = 54.8  # 1 kg/yr ≈ 2.74 g/day × 20 mg/g
beverage['est_caffeine_mg_day'] = (
    beverage['coffee_kg_per_capita'] * COFFEE_FACTOR +
    beverage['tea_kg_per_capita']    * TEA_FACTOR)
beverage['dominant_source'] = np.where(
    beverage['coffee_kg_per_capita'] > beverage['tea_kg_per_capita'],
    'Coffee-dominant', 'Tea-dominant')
print(f"Beverage cross-section: {beverage.shape[0]} countries")

merged = panel.merge(beverage, on='country', how='inner')
print(f"\nMerged panel: {merged.shape[0]} rows, "
      f"{merged['country'].nunique()} countries, "
      f"years {merged['year'].min():.0f}–{merged['year'].max():.0f}")

if 'depression_pct' in merged.columns and merged['depression_pct'].notna().any():
    RECENT_YEAR = int(merged.dropna(subset=['depression_pct'])['year'].max())
elif 'mortality_per_100k' in merged.columns and merged['mortality_per_100k'].notna().any():
    RECENT_YEAR = int(merged.dropna(subset=['mortality_per_100k'])['year'].max())
else:
    RECENT_YEAR = int(merged['year'].max())

cross_section = merged[merged['year'] == RECENT_YEAR].copy()
print(f"Most recent cross-section (year={RECENT_YEAR}): {cross_section.shape[0]} countries")

print(f"\nMissing values in full panel:")
display(merged.isnull().sum().to_frame('missing').assign(
    pct=lambda x: (x['missing'] / len(merged) * 100).round(1)))

PROCESSED.mkdir(parents=True, exist_ok=True)
merged.to_csv(PROCESSED / 'merged_panel.csv', index=False)
cross_section.to_csv(PROCESSED / 'merged_cross_section.csv', index=False)
```

#### Section 1: Temporal Coverage and Outcome Trends

Before examining caffeine–health associations, we need to understand the temporal structure of our panel: which years have good coverage, how outcomes are trending globally, and whether trends differ for high-caffeine vs. low-caffeine countries.

#### Section 3: Caffeine Exposure and Health Outcomes

This section examines the main relationship in our project. We look at the caffeine–health association from three angles:
1. **Cross-sectional scatter plots** (most recent year) with correlation statistics
2. **Correlation stability over time** — does the association appear across multiple years?
3. **Multi-year scatter grid** — a visual check of temporal consistency


```python
## Reusable scatter plot function (also in modules/eda_utils.py)
def labeled_scatter(df, x, y, xlabel, ylabel, title,
                    label_col='country', n_labels=5, ax=None):
    """Scatter + OLS line + Pearson/Spearman stats + extreme-point labels."""
    if ax is None:
        fig, ax = plt.subplots(figsize=(9, 6))
    subset = df[[x, y, label_col]].dropna()
    if len(subset) < 5:
        ax.text(0.5, 0.5, 'Insufficient data', transform=ax.transAxes, ha='center')
        ax.set_title(title)
        return ax
    sns.regplot(x=x, y=y, data=subset, ax=ax,
                scatter_kws={'alpha': 0.5, 's': 50, 'color': 'steelblue'},
                line_kws={'color': 'crimson', 'lw': 2})
    r_p, p_p = stats.pearsonr(subset[x], subset[y])
    r_s, p_s = stats.spearmanr(subset[x], subset[y])
    ax.text(0.03, 0.97,
            f'Pearson r = {r_p:.3f} (p = {p_p:.3g})\n'
            f'Spearman ρ = {r_s:.3f} (p = {p_s:.3g})',
            transform=ax.transAxes, va='top', fontsize=10,
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    for _, row in subset.nlargest(n_labels, x).iterrows():
        ax.annotate(row[label_col], (row[x], row[y]),
                    fontsize=8, alpha=0.7, xytext=(5, 5), textcoords='offset points')
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    return ax
```

```python
## Cross-sectional scatter plots (most recent year)
outcome_plots = []
if 'depression_pct' in cross_section.columns:
    outcome_plots.append(('est_caffeine_mg_day', 'depression_pct',
        'Estimated Caffeine (mg/day)', 'Depression Prevalence (%)',
        f'Caffeine vs. Depression ({RECENT_YEAR})'))
if 'mortality_per_100k' in cross_section.columns:
    outcome_plots.append(('est_caffeine_mg_day', 'mortality_per_100k',
        'Estimated Caffeine (mg/day)', 'All-Cause Mortality (per 100k)',
        f'Caffeine vs. Mortality ({RECENT_YEAR})'))

if outcome_plots:
    fig, axes = plt.subplots(1, len(outcome_plots), figsize=(8*len(outcome_plots), 6))
    if len(outcome_plots) == 1: axes = [axes]
    for ax, (x, y, xl, yl, title) in zip(axes, outcome_plots):
        labeled_scatter(cross_section, x, y, xl, yl, title, ax=ax)
    plt.tight_layout()
    plt.savefig(FIG / '03a_caffeine_vs_outcomes.png', dpi=150, bbox_inches='tight')
    plt.show()
else:
    print("No outcome columns available for scatter plots.")
```

```python
## How does the caffeine-outcome correlation change over time?
corr_over_time = []
for year in sorted(merged['year'].dropna().unique()):
    yr_data = merged[merged['year'] == year]
    if len(yr_data) < 20:
        continue
    row = {'year': int(year)}
    for col, label in outcome_vars[:2]:
        subset = yr_data[['est_caffeine_mg_day', col]].dropna()
        if len(subset) >= 10:
            r, p = stats.spearmanr(subset['est_caffeine_mg_day'], subset[col])
            row[f'{label}_rho'] = r
            row[f'{label}_p'] = p
            row[f'{label}_n'] = len(subset)
    corr_over_time.append(row)

corr_df = pd.DataFrame(corr_over_time)

if len(corr_df) > 0:
    fig, ax = plt.subplots(figsize=(12, 5))
    for col, label in outcome_vars[:2]:
        rho_col = f'{label}_rho'
        if rho_col in corr_df.columns:
            ax.plot(corr_df['year'], corr_df[rho_col],
                    marker='o', markersize=4, lw=1.5, label=f'Caffeine vs. {label}')
    ax.axhline(0, color='gray', ls='--', alpha=0.5)
    ax.set_xlabel('Year')
    ax.set_ylabel('Spearman ρ')
    ax.set_title('Stability of Caffeine–Outcome Correlations Over Time\n'
                 '(cross-sectional Spearman rank correlation, computed per year)')
    ax.legend()
    ax.set_ylim(-1, 1)
    plt.tight_layout()
    plt.savefig(FIG / '03b_correlation_stability.png', dpi=150, bbox_inches='tight')
    plt.show()
else:
    print("Not enough data to compute correlations over time.")
```

```python
## Scatter plots at different time points
primary_outcome = outcome_vars[0][0] if outcome_vars else None
primary_label = outcome_vars[0][1] if outcome_vars else None

if primary_outcome and primary_outcome in merged.columns:
    sample_years = [y for y in sorted(merged['year'].unique())
                    if y % 5 == 0 and merged[merged['year']==y].dropna(
                        subset=['est_caffeine_mg_day', primary_outcome]).shape[0] >= 15]
    if len(sample_years) > 4:
        sample_years = sample_years[-4:]

    if sample_years:
        fig, axes = plt.subplots(1, len(sample_years),
                                 figsize=(5*len(sample_years), 5), sharey=True)
        if len(sample_years) == 1: axes = [axes]

        for ax, yr in zip(axes, sample_years):
            yr_data = merged[merged['year'] == yr].dropna(
                subset=['est_caffeine_mg_day', primary_outcome])
            sns.regplot(x='est_caffeine_mg_day', y=primary_outcome, data=yr_data,
                        ax=ax, scatter_kws={'alpha': 0.4, 's': 30, 'color': 'steelblue'},
                        line_kws={'color': 'crimson', 'lw': 1.5})
            if len(yr_data) >= 5:
                r, p = stats.spearmanr(yr_data['est_caffeine_mg_day'], yr_data[primary_outcome])
                ax.set_title(f'{int(yr)}\nρ = {r:.2f} (p = {p:.3f})')
            else:
                ax.set_title(f'{int(yr)}\n(n={len(yr_data)})')
            ax.set_xlabel('Caffeine (mg/day)')
            if ax == axes[0]:
                ax.set_ylabel(f'{primary_label}')

        plt.suptitle(f'Caffeine vs. {primary_label}: Cross-Sections Over Time',
                     fontsize=14, y=1.03)
        plt.tight_layout()
        plt.savefig(FIG / '03c_scatter_grid.png', dpi=150, bbox_inches='tight')
        plt.show()
    else:
        print("Not enough years with sufficient data for scatter grid.")
```

**Interpretation:**

The cross-sectional scatter plots show the direction and strength of the caffeine–health association in the most recent available year. Spearman correlations are reported alongside Pearson to reduce the effect of skewness.

The correlation-over-time plot shows whether the caffeine–depression and caffeine–mortality associations appear repeatedly across years. A pattern that appears in multiple years is more informative than a single snapshot, although it still does not establish causation.

The multi-year scatter grid provides a visual check of whether the relationship looks similar at different time points.

**Caveat:** These are still ecological correlations. Any negative association between caffeine and mortality could still be explained by GDP or other confounders, which is why the next section focuses on confounder structure.

## Ethics

Our project uses publicly available, aggregated country-level data, which minimizes individual privacy concerns. However, several ethical considerations apply:

**Ecological fallacy risk:** Our analysis operates at the country level. It would be irresponsible to use these findings to make claims about individual caffeine consumption and health. We must be transparent that country-level correlations do not imply individual-level causation.

**Cultural sensitivity:** Caffeine consumption patterns are deeply tied to culture (e.g., tea ceremonies in East Asia, coffee culture in Scandinavia). As well in Turkey, The Middle East, and Asia, where there are high-volume of traditional and social rituals with caffeine. Framing any pattern as "better" or "worse" based on caffeine source could reinforce cultural biases.

**Data representativeness:** Model-based estimates from IHME/GBD may be less reliable for low-income countries with weaker health surveillance systems, potentially biasing results toward patterns observed in well-studied, high-income nations.

**Potential for misuse:** Findings could be taken out of context by the caffeine/beverage industry for marketing purposes. We emphasize that this is an observational, ecological study and no causal claims should be derived from it.
