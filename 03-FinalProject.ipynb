## Abstract

This project investigates whether country-level estimated caffeine consumption per capita is associated with national mental health and mortality outcomes, after accounting for major economic and lifestyle confounders. We constructed a country-year panel by merging per-capita coffee and tea consumption data (converted to estimated daily caffeine in mg/person/day) with depression prevalence and all-cause mortality from IHME GBD, alongside GDP per capita, smoking prevalence, and alcohol consumption from Our World in Data.

The caffeine-depression association was weaker and less consistent across years.

# Overview

## Research Question

We will test whether **estimated caffeine consumption per capita (by country)** is associated with **country-level health and social outcomes**, after accounting for major confounders.

Primary question: Across countries, is higher estimated caffeine intake associated with differences in mental health and mortality indicators (e.g., depression burden and all-cause mortality)?

Secondary questions:
- Do associations differ depending on the dominant caffeine source (coffee vs tea vs carbonated/energy drinks)?
- Do associations remain after controlling for factors like GDP per capita, smoking prevalence, alcohol use, and other relevant confounders?

This is an **inference** project using regression-based modeling and sensitivity analyses to test robustness to uncertainty in the caffeine estimates.

## Background and Prior Work

Caffeine is one of the most widely consumed psychoactive substances worldwide, typically consumed through coffee, tea, and other caffeinated beverages. Because caffeine affects sleep, mood, and cardiovascular physiology, population-level differences in caffeine intake could plausibly relate to mental health and mortality patterns.

At the individual level, observational research has reported associations between caffeine intake and depression-related outcomes. For example, analyses using NHANES data examined caffeine consumption and depression measures, suggesting a relationship may exist, though causality is not established and confounding remains a major concern.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Similarly, large observational analyses have examined caffeine intake and all-cause or cause-specific mortality, often identifying non-linear associations and emphasizing careful interpretation and adjustment for confounders.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

However, most prior work is within-country and individual-level, while our project is between-country and ecological. This makes the question different: we are not claiming caffeine causes outcomes in individuals, but asking whether country-level patterns co-vary and whether those patterns persist after controlling for structural factors (wealth, health system capacity, smoking, alcohol use, etc.).

To estimate country-level caffeine intake, we start from publicly available per-capita consumption estimates for coffee and tea by country for recent years.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) We then convert these amounts to approximate caffeine intake using caffeine-content references (acknowledging large variation across preparation methods and products).<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5)

In addition, beverage-market composition data (percent of caffeinated beverage volume sales by category) can help characterize where caffeine is coming from in different countries (coffee vs tea vs carbonates vs energy drinks), which we use for subgroup and interaction analyses.<a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6)

## Hypothesis

We hypothesize that countries with higher estimated caffeine consumption per capita will show measurable differences in mental health and mortality indicators, though the direction may differ by outcome. Based on prior observational findings, we predict a weak-to-moderate association between higher caffeine intake and lower depression burden and/or lower all-cause mortality, but we expect substantial confounding by economic development and lifestyle factors.<a name="cite_ref-1b"></a>[<sup>1</sup>](#cite_note-1)

We also hypothesize that the dominant caffeine source will matter: countries where caffeine comes mainly from energy drinks or carbonated beverages may show different patterns than countries where caffeine primarily comes from coffee or tea.<a name="cite_ref-6b"></a>[<sup>6</sup>](#cite_note-6)

### Regression Analysis

We fit a series of OLS regression models to test whether estimated caffeine intake is associated with depression prevalence and all-cause mortality after controlling for confounders. Three model specifications are run for each outcome:

- **Model 1 (Unadjusted):** `outcome ~ caffeine`
- **Model 2 (GDP-adjusted):** `outcome ~ caffeine + log(GDP per capita)`
- **Model 3 (Fully adjusted):** `outcome ~ caffeine + log(GDP per capita) + smoking + alcohol`

GDP is log-transformed due to right skew. All continuous predictors are z-scored so that coefficients are comparable in magnitude. The primary analysis uses the most recent cross-section. A sensitivity analysis varies the caffeine conversion factors by plus or minus 25% to check how sensitive results are to our approximation assumptions.

```python
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Prepare cross-section for regression
reg_cols = ['country', 'est_caffeine_mg_day', 'depression_pct', 'mortality_per_100k',
            'gdp_per_capita', 'smoking_pct', 'alcohol_liters']
reg_cols_present = [c for c in reg_cols if c in cross_section.columns]
reg_df = cross_section[reg_cols_present].dropna().copy()

# Log-transform GDP, z-score all continuous predictors
reg_df['log_gdp'] = np.log(reg_df['gdp_per_capita'])

z_cols = ['est_caffeine_mg_day', 'log_gdp', 'smoking_pct', 'alcohol_liters']
z_cols = [c for c in z_cols if c in reg_df.columns]
for col in z_cols:
    reg_df[f'{col}_z'] = (reg_df[col] - reg_df[col].mean()) / reg_df[col].std()

print(f"Regression sample: {len(reg_df)} countries")
print(f"Predictors available: {z_cols}")
display(reg_df[z_cols + [c for c in ['depression_pct', 'mortality_per_100k'] if c in reg_df.columns]]
        .describe().round(2))
```

```python
# VIF check for multicollinearity
vif_cols = [f'{c}_z' for c in z_cols if f'{c}_z' in reg_df.columns]
vif_data = reg_df[vif_cols].dropna()

vif_results = pd.DataFrame({
    'Variable': vif_cols,
    'VIF': [variance_inflation_factor(vif_data.values, i)
            for i in range(len(vif_cols))]
}).round(2)

print("Variance Inflation Factors (VIF > 5 = multicollinearity concern):")
display(vif_results)
```

```python
# OLS regression: Depression outcome
results_depression = {}
outcome = 'depression_pct'

if outcome in reg_df.columns:
    results_depression['Model 1 (Unadjusted)'] = smf.ols(
        f'{outcome} ~ est_caffeine_mg_day_z', data=reg_df).fit()

    if 'log_gdp_z' in reg_df.columns:
        results_depression['Model 2 (GDP-adjusted)'] = smf.ols(
            f'{outcome} ~ est_caffeine_mg_day_z + log_gdp_z', data=reg_df).fit()

    add_controls = [f'{c}_z' for c in ['log_gdp', 'smoking_pct', 'alcohol_liters']
                    if f'{c}_z' in reg_df.columns]
    if add_controls:
        results_depression['Model 3 (Fully adjusted)'] = smf.ols(
            f'{outcome} ~ est_caffeine_mg_day_z + {" + ".join(add_controls)}',
            data=reg_df).fit()

    for name, res in results_depression.items():
        print(f"\n{'='*60}")
        print(f"{name} | Outcome: {outcome} | N={int(res.nobs)} | R2={res.rsquared:.3f} | Adj.R2={res.rsquared_adj:.3f}")
        print(res.summary2().tables[1].round(3))
else:
    print(f"'{outcome}' not found in dataset.")
```

```python
# OLS regression: Mortality outcome
results_mortality = {}
outcome = 'mortality_per_100k'

if outcome in reg_df.columns:
    results_mortality['Model 1 (Unadjusted)'] = smf.ols(
        f'{outcome} ~ est_caffeine_mg_day_z', data=reg_df).fit()

    if 'log_gdp_z' in reg_df.columns:
        results_mortality['Model 2 (GDP-adjusted)'] = smf.ols(
            f'{outcome} ~ est_caffeine_mg_day_z + log_gdp_z', data=reg_df).fit()

    add_controls = [f'{c}_z' for c in ['log_gdp', 'smoking_pct', 'alcohol_liters']
                    if f'{c}_z' in reg_df.columns]
    if add_controls:
        results_mortality['Model 3 (Fully adjusted)'] = smf.ols(
            f'{outcome} ~ est_caffeine_mg_day_z + {" + ".join(add_controls)}',
            data=reg_df).fit()

    for name, res in results_mortality.items():
        print(f"\n{'='*60}")
        print(f"{name} | Outcome: {outcome} | N={int(res.nobs)} | R2={res.rsquared:.3f} | Adj.R2={res.rsquared_adj:.3f}")
        print(res.summary2().tables[1].round(3))
else:
    print(f"'{outcome}' not found in dataset.")
```

```python
# Coefficient plot: caffeine coefficient across all models and outcomes
all_results = {}
all_results.update({f'Depression | {k}': v for k, v in results_depression.items()})
all_results.update({f'Mortality  | {k}': v for k, v in results_mortality.items()})

coef_rows = []
for label, res in all_results.items():
    if 'est_caffeine_mg_day_z' in res.params:
        coef = res.params['est_caffeine_mg_day_z']
        se   = res.bse['est_caffeine_mg_day_z']
        pval = res.pvalues['est_caffeine_mg_day_z']
        coef_rows.append({'Model': label, 'coef': coef, 'se': se, 'pval': pval,
                          'ci_lo': coef - 1.96*se, 'ci_hi': coef + 1.96*se})

coef_df = pd.DataFrame(coef_rows)

if len(coef_df) > 0:
    fig, ax = plt.subplots(figsize=(9, 0.7 * len(coef_df) + 2))
    y_pos  = range(len(coef_df))
    colors = ['#e74c3c' if p < 0.05 else '#95a5a6' for p in coef_df['pval']]
    ax.barh(y_pos, coef_df['coef'], xerr=1.96*coef_df['se'],
            color=colors, alpha=0.8, capsize=4, height=0.5)
    ax.axvline(0, color='black', lw=1, ls='--')
    ax.set_yticks(y_pos)
    ax.set_yticklabels(coef_df['Model'], fontsize=9)
    ax.set_xlabel('Caffeine coefficient (standardized) +/- 95% CI')
    ax.set_title('Caffeine Coefficient Across Models and Outcomes\n'
                 '(red = p < 0.05, gray = not significant)')
    plt.tight_layout()
    plt.savefig(FIG / '05a_coef_plot.png', dpi=150, bbox_inches='tight')
    plt.show()

    print("\nCaffeine coefficients summary:")
    display(coef_df[['Model', 'coef', 'se', 'pval']].round(4))
```

```python
# Residual diagnostics for fully adjusted models
for outcome_label, res_dict in [('Depression', results_depression), ('Mortality', results_mortality)]:
    if not res_dict:
        continue
    key = 'Model 3 (Fully adjusted)' if 'Model 3 (Fully adjusted)' in res_dict else list(res_dict.keys())[-1]
    res       = res_dict[key]
    fitted    = res.fittedvalues
    residuals = res.resid

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    axes[0].scatter(fitted, residuals, alpha=0.5, color='steelblue', s=40)
    axes[0].axhline(0, color='crimson', ls='--', lw=1.5)
    axes[0].set_xlabel('Fitted values')
    axes[0].set_ylabel('Residuals')
    axes[0].set_title(f'{outcome_label}: Residuals vs Fitted')

    stats.probplot(residuals, dist='norm', plot=axes[1])
    axes[1].set_title(f'{outcome_label}: Normal Q-Q Plot of Residuals')

    plt.suptitle(f'{key}', fontsize=12, y=1.02)
    plt.tight_layout()
    plt.savefig(FIG / f'05b_diagnostics_{outcome_label.lower()}.png', dpi=150, bbox_inches='tight')
    plt.show()
```

```python
# Sensitivity analysis: vary caffeine conversion factors by +/- 25%
COFFEE_FACTOR_DEFAULT = 32.9
TEA_FACTOR_DEFAULT    = 54.8

scenarios = {
    'Low estimate (-25%)':  (COFFEE_FACTOR_DEFAULT * 0.75, TEA_FACTOR_DEFAULT * 0.75),
    'Default':              (COFFEE_FACTOR_DEFAULT,         TEA_FACTOR_DEFAULT),
    'High estimate (+25%)': (COFFEE_FACTOR_DEFAULT * 1.25, TEA_FACTOR_DEFAULT * 1.25),
}

sensitivity_rows = []
for scenario_name, (cf, tf) in scenarios.items():
    tmp = reg_df.copy()
    if 'coffee_kg_per_capita' in tmp.columns and 'tea_kg_per_capita' in tmp.columns:
        tmp['caffeine_s'] = tmp['coffee_kg_per_capita'] * cf + tmp['tea_kg_per_capita'] * tf
    else:
        scale = cf / COFFEE_FACTOR_DEFAULT
        tmp['caffeine_s'] = tmp['est_caffeine_mg_day'] * scale

    tmp['caffeine_s_z'] = (tmp['caffeine_s'] - tmp['caffeine_s'].mean()) / tmp['caffeine_s'].std()

    for outcome in ['depression_pct', 'mortality_per_100k']:
        if outcome not in tmp.columns:
            continue
        add_controls = [f'{c}_z' for c in ['log_gdp', 'smoking_pct', 'alcohol_liters']
                        if f'{c}_z' in tmp.columns]
        formula = f'{outcome} ~ caffeine_s_z + {" + ".join(add_controls)}'
        try:
            fit  = smf.ols(formula, data=tmp).fit()
            coef = fit.params['caffeine_s_z']
            pval = fit.pvalues['caffeine_s_z']
            sensitivity_rows.append({'Scenario': scenario_name, 'Outcome': outcome,
                                     'Caffeine coef': round(coef, 4), 'p-value': round(pval, 4)})
        except Exception as e:
            print(f"Skipped {scenario_name}/{outcome}: {e}")

if sensitivity_rows:
    sens_df = pd.DataFrame(sensitivity_rows)
    print("Sensitivity Analysis: Caffeine coefficient under +/-25% conversion factor variation")
    display(sens_df.pivot(index='Outcome', columns='Scenario', values=['Caffeine coef', 'p-value']))
else:
    print("Sensitivity analysis could not run. Check that coffee/tea columns are available.")
```

## Ethics

Our project uses publicly available, aggregated country-level data, which minimizes individual privacy concerns. However, several ethical considerations apply.

**Ecological fallacy risk.** Our analysis operates at the country level. It would be irresponsible to make use of these findings in order to make claims about individual caffeine consumption and health. A country with high average caffeine intake might have better health outcomes for reasons entirely unrelated to caffeine, and our data cannot rule that out. We have been explicit throughout that country-level correlations do not imply individual-level causation.

**Cultural sensitivity.** Caffeine consumption patterns are deeply tied to culture and history — tea ceremonies in East Asia, coffee culture in Scandinavia, traditional qahwa in the Middle East and Turkey. Framing any observed pattern as better or worse based on caffeine source or quantity risks being culturally reductive. Our analysis describes co-variation across countries, not prescriptions about which consumption patterns are desirable.

**Data representativeness.** Model-based estimates from IHME/GBD may be less reliable for low-income countries with weaker health surveillance systems, which could bias results toward patterns observed in well-studied, high-income nations. High-caffeine countries in our dataset are also the most accurately measured countries, which introduces a systematic asymmetry that is difficult to correct for.

**Potential for misuse.** Findings from ecological studies like this can be selectively cited by the beverage industry for marketing purposes. We emphasize that this is an observational, ecological study and no causal claims are warranted by the design.

## Discussion and Conclusion

This project asked whether country-level estimated caffeine consumption is associated with national mental health and mortality outcomes, and whether any such association persists after controlling for economic development and lifestyle factors.

**What the EDA showed.** The exploratory analysis found a consistent negative correlation between the estimation of caffeine intake and the all-cause mortality across countries and across multiple years in the panel. Countries in the highest caffeine tercile showed lower median mortality throughout. The caffeine-depression association was weaker and less stable over time. These patterns are consistent in direction with prior individual-level findings, though the ecological level of analysis makes direct comparison difficult.

**The confounding problem.** The most important finding from the EDA may be the confounder structure itself. GDP per capita is strongly correlated with both caffeine intake and mortality, which means the raw caffeine-mortality correlation is almost certainly partially a wealth effect. The GDP-stratified scatter plots showed weaker within-stratum associations compared to the overall association, which motivated the regression approach.

**Extending prior work.** Most prior caffeine-health research is individual-level and within a single country. Our ecological approach is complementary but distinct. We are not asking whether individuals who drink more coffee are healthier, but whether structural patterns of caffeine use at the national level track with population health outcomes. An ecological association could exist even if the individual-level effect were null, and vice versa. This is why we have not generalized our findings to individual-level claims.

**Limitations.** Several limitations bear emphasis. First, the caffeine exposure variable is approximate. Caffeine content varies substantially by preparation method and product, and we used fixed conversion factors applied uniformly across all countries. Second, coffee and tea consumption inputs are single cross-sections applied to a multi-year panel, which assumes stability over time. Third, the panel structure means our caffeine measure from 2022-2023 is being used to explain outcomes across many earlier years, introducing temporal misalignment. Fourth, IHME/GBD estimates are more reliable for wealthier countries, which are also the high-caffeine countries in our sample, potentially inflating observed associations. Fifth, we did not include energy drink consumption, which is growing rapidly in many middle-income countries and may not track with coffee and tea patterns.

**Next steps.** If this work continued, useful extensions would include obtaining energy drink market share data at the country level, running country fixed-effects models to remove time-invariant national characteristics rather than just controlling for GDP cross-sectionally, and acquiring year-matched beverage consumption data rather than a single cross-section.

**Bottom line.** Country-level caffeine consumption shows a negative association with all-cause mortality that is at least partially, and possibly not entirely, explained by economic development. The depression association is weaker and less consistent. These findings are consistent in direction with prior individual-level literature, but the ecological design and approximate exposure measure mean they should be interpreted as suggestive rather than confirmatory.
