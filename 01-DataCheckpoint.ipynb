## Research Question

We will test whether **estimated caffeine consumption per capita (by country)** is associated with **country-level health and social outcomes**, after accounting for major confounders.

Primary question: Across countries, is higher estimated caffeine intake associated with differences in mental health and mortality indicators (e.g., depression burden and all-cause mortality)?

Secondary questions:
- Do associations differ depending on the dominant caffeine source (coffee vs tea vs carbonated/energy drinks)?
- Do associations remain after controlling for factors like GDP per capita, smoking prevalence, alcohol use, and other relevant confounders?

This is an **inference** project using regression-based modeling and sensitivity analyses to test robustness to uncertainty in the caffeine estimates.



## Background and Prior Work

Caffeine is one of the most widely consumed psychoactive substances worldwide, typically consumed through coffee, tea, and other caffeinated beverages. Because caffeine affects sleep, mood, and cardiovascular physiology, population-level differences in caffeine intake could plausibly relate to mental health and mortality patterns.

At the individual level, observational research has reported associations between caffeine intake and depression-related outcomes. For example, analyses using NHANES data examined caffeine consumption and depression measures, suggesting a relationship may exist, though causality is not established and confounding remains a major concern.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Similarly, large observational analyses have examined caffeine intake and all-cause or cause-specific mortality, often identifying non-linear associations and emphasizing careful interpretation and adjustment for confounders.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

However, most prior work is within-country and individual-level, while our project is between-country and ecological. This makes the question different: we are not claiming caffeine causes outcomes in individuals, but asking whether country-level patterns co-vary and whether those patterns persist after controlling for structural factors (wealth, health system capacity, smoking, alcohol use, etc.).

To estimate country-level caffeine intake, we will start from publicly available per-capita consumption estimates for coffee and tea by country for recent years.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) We will then convert these amounts to approximate caffeine intake using caffeine-content references (acknowledging large variation across preparation methods and products).<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5)

In addition, beverage-market composition data (percent of caffeinated beverage volume sales by category) can help characterize where caffeine is coming from in different countries (coffee vs tea vs carbonates vs energy drinks), which we can use for subgroup and interaction analyses.<a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6)

Based on prior observational findings, we predict a weak-to-moderate association between higher caffeine intake and lower depression burden and/or lower all-cause mortality, but we expect substantial confounding by economic development and lifestyle factors.<a name="cite_ref-1b"></a>[<sup>1</sup>](#cite_note-1)

We also hypothesize that the dominant caffeine source will matter: countries where caffeine comes mainly from energy drinks or carbonated beverages may show different patterns than countries where caffeine primarily comes from coffee or tea.<a name="cite_ref-6b"></a>[<sup>6</sup>](#cite_note-6)

# Coffee Consumption per Capita by Country (2023)

This dataset reports **country-level coffee consumption rates** measured in **kilograms of coffee consumed per person per year (kg/capita)** for **2023**. Each row corresponds to a country and the main metric of interest is the *Coffee Consumption Rate 2023*, which can be interpreted as an estimate of the average amount of coffee consumed per resident in that country during the year.

In practical terms, higher values indicate populations where coffee is consumed more heavily (e.g., strong coffee culture and/or high availability), while lower values indicate less coffee consumption on average. Because this is a **per-capita national aggregate**, it does **not** tell us about individual consumption patterns, within-country inequality (some people drink a lot while others drink none), or *caffeine* intake specifically (different coffee products and brewing methods can deliver very different caffeine doses).

Major concerns/shortcomings:

- **Ecological / aggregate data**: country averages can’t be used to infer individual behavior (ecological fallacy).
- **Comparability & measurement error**: definitions may differ across sources/countries (what counts as “coffee consumption”), reporting methods vary, and estimates may include multiple coffee product categories.
- **Not direct caffeine measurement**: kg of coffee does not translate cleanly to mg caffeine without assumptions about bean type, brewing, serving sizes, etc.


```python
## Load, tidy, inspect, and clean the dataset

# A) Load the dataset (from the provided link) and save a raw copy
url = 'https://worldpopulationreview.com/country-rankings/coffee-consumption-by-country'
raw_html_path = RAW_DIR / 'coffee_consumption_by_country_2023.html'

# Download the webpage as our "raw" data snapshot
resp = requests.get(url, timeout=30)
resp.raise_for_status()
raw_html_path.write_text(resp.text, encoding='utf-8')

# Parse tables from the HTML snapshot
tables = pd.read_html(raw_html_path)

# Find the table that contains the expected columns
df = None
for t in tables:
    cols = [str(c).lower() for c in t.columns]
    if any('country' == c for c in cols) and any('coffee' in c and '2023' in c for c in cols):
        df = t.copy()
        break

if df is None:
    # fallback: just use the first table and inspect
    df = tables[0].copy()

# Standardize column names
df.columns = [
    re.sub(r'[^0-9a-zA-Z]+', '_', str(c).strip()).strip('_').lower()
    for c in df.columns
]

# Keep only the key variables (Country + 2023 consumption rate) if present
country_col = 'country' if 'country' in df.columns else df.columns[0]
rate_candidates = [c for c in df.columns if ('coffee' in c and '2023' in c) or ('consumption' in c and '2023' in c)]
rate_col = rate_candidates[0] if rate_candidates else df.columns[1]

df = df[[country_col, rate_col]].rename(columns={country_col: 'country', rate_col: 'coffee_kg_per_capita_2023'})

# B) Tidy check: already tidy (one row per country, one variable per column)

# C) Demonstrate the size of the dataset
print('Rows, Columns:', df.shape)

# D) Missingness
missing = df.isna().sum().to_frame('missing_count')
missing['missing_pct'] = missing['missing_count'] / len(df)
display(missing)

# Convert the rate to numeric (coerce bad strings to NaN)
df['coffee_kg_per_capita_2023'] = pd.to_numeric(df['coffee_kg_per_capita_2023'], errors='coerce')

# Re-check missingness after coercion
print('\nMissing after numeric conversion:')
display(df.isna().sum().to_frame('missing_count'))

# E) Flag potential outliers / suspicious entries
desc = df['coffee_kg_per_capita_2023'].describe()
print('\nSummary of coffee_kg_per_capita_2023:')
print(desc)

# Simple outlier flag using IQR rule (not deleting automatically)
q1 = df['coffee_kg_per_capita_2023'].quantile(0.25)
q3 = df['coffee_kg_per_capita_2023'].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

df['outlier_iqr'] = (df['coffee_kg_per_capita_2023'] < lower) | (df['coffee_kg_per_capita_2023'] > upper)
print('\nNumber of IQR-flagged outliers:', int(df['outlier_iqr'].sum()))
display(df.sort_values('coffee_kg_per_capita_2023', ascending=False).head(10))

# F) Clean: drop rows missing key variables (country or rate)
# Justification: without the 2023 rate, the row is not usable for analysis.
df_clean = df.dropna(subset=['country', 'coffee_kg_per_capita_2023']).copy()

# Remove duplicates by country (keep first) if any
df_clean = df_clean.drop_duplicates(subset=['country'])

print('\nCleaned rows, columns:', df_clean.shape)

# G) Save intermediate and processed versions
interim_path = INTERIM_DIR / 'coffee_consumption_by_country_2023_interim.csv'
processed_path = PROCESSED_DIR / 'coffee_consumption_by_country_2023_processed.csv'

df.to_csv(interim_path, index=False)
df_clean.to_csv(processed_path, index=False)

# Optional: show a histogram for quick EDA
ax = sns.histplot(df_clean['coffee_kg_per_capita_2023'], bins=25)
ax.set_xlabel('Coffee consumption (kg per capita, 2023)')
ax.set_ylabel('Count of countries')
plt.show()
```
